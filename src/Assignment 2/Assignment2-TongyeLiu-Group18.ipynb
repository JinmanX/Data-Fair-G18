```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("./conflictlocation.csv")
#read the file first
data.head()
```

```python
#I found that the 'tweet_text' contains more contents than others, so change the colwidth to observe
pd.set_option('max_colwidth', 300)
data
```

```python
# Tweet text begin with b' and it need to be removed
# There are x/.., mentions,hashtags and links in tweet_text, I decide to remove them because they were useless
# Import re to implement matching and replacing in string

import re
# Remove x/.. and b', and then check
data['tweet_text'] = data['tweet_text'].apply(lambda x:re.sub(r'\\x..',"",x.replace("b'","")))
data.head()
```

```python
# Remove links and check 
data['tweet_text'] = data['tweet_text'].apply(lambda x:re.sub(r"((https|ftp|file)://[-A-Za-z0-9+&@#/%?=~_|!:,.;]+[-A-Za-z0-9+&@#/%=~_|])","",x))
data.head()
```

```python
# Remove mentions and hashtags and check
data['tweet_text'] = data['tweet_text'].apply(lambda x:re.sub(r"#(\w+)","",x))
data['tweet_text'] = data['tweet_text'].apply(lambda x:re.sub(r"@(\w+)","",x))
data.head()
```

```python
# Tweet_text is clean now. Next step is to remove "b'" from the username and check
data['username'] = data['username'].apply(lambda x:x[2:-1])
data.head()
```

```python
# The hashtags need to be strings for analysis
# Turn all_hashtags to a list through eval, and then turn it into a string and check
data['all_hashtags'] = data['all_hashtags'].apply(lambda x:' '.join(eval(x)))
data.head()
```

```python
# Then the analysis shall begin
# I'll go with the timestamp first, I want to know in which time period there are the most tweets
# I think it may when having meal and break
# So I draw a histogram to find out
data['timestamp'] = pd.to_datetime(data['timestamp'])
times = data['timestamp'].sort_values()
plt.figure(figsize=(10,8),dpi=80)
plt.hist(times,bins=30)
plt.savefig('./times.png')
# It turns out that people like to post in 9am, 4pm and 1pm, which may be breakfast time, lunch time and afternoon break.
```

```python
# Then about the location, there are some tweets without location. I want to find out the percentage of people with location.
# And in what place people tweets most in these days.
total_row = data.shape[0]
location_null_row = data['location'].isnull().sum()
# Calculate the total number of rows and the number of rows with empty locations.
plt.figure(figsize=(20,8),dpi=80)
plt.pie([location_null_row,total_row-location_null_row],labels=['Null','Not Null'],autopct='%1.1f%%')
plt.title('Analyze Result of Location',fontsize=10)
plt.savefig("././locations.png")
# Calculate the percentage and draw a pie chart of it
```

```python
# Filter out the non-empty value in the location columnï¼Œthen count through the collections library
import collections
locationGroup = collections.Counter(data[data['location'].notnull()]['location'])
# Find the top 10 places that occur the most, and draw a chart
mostCommonLocation = locationGroup.most_common(10)
locations = [x[0] for x in mostCommonLocation]
locationCount = [x[1] for x in mostCommonLocation]
plt.figure(figsize=(20,8),dpi=80)
plt.bar(locations,locationCount)
plt.xlabel('Location',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Most Common Location',fontsize=20)
plt.savefig("./common_locations.png")
```

```python
# Next column is tweet_text, and I decide to do sentimental analysis on it
# I serched and found vader which is a sentiment analysis tool based on vocabulary and rules
# It is particularly suitable for the sentiment expressed in social media, and also has good applications in texts in other fields
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
vader = SentimentIntensityAnalyzer()
senti_result = {
    'negative':0,
    'positive':0,
    'neutral':0
}
```

```python
# Use math to classify and count the tweets according to emotion intensity
import math
for each in data['tweet_text']:
    each_result = vader.polarity_scores(each)
    compound = each_result.get('compound')
    if math.fabs(compound) <= 0.2:
        senti_result['neutral']+=1
    elif compound > 0:
        senti_result['positive']+=1
    else:
        senti_result['negative']+=1
```

```python
keys = list()
values = list()
for each in senti_result:
    keys.append(each)
    values.append(senti_result[each])
```

```python
plt.figure(figsize=(20,8),dpi=80)
plt.pie(values,labels=keys,autopct='%1.1f%%')
plt.title('Sentimental Analysis Result',fontsize=10)
plt.savefig("././sentimental.png")
# Most of the tweets are judged as positive.
# I'm not sure if this means people are more neutral and positive when describe conflict.
```

```python
import wordcloud
#First remove 'conflict' because every row contains it
#Then count the word frequency through the collections library
# Draw a frequency distribution diagram
words = " ".join(data['all_hashtags']).lower().replace('conflict','')
freq = collections.Counter([x for x in words.split(" ") if len(x) > 0])
plt.hist(freq.values())
plt.show()
```

```python
#Generate the wordcloud
wc = wordcloud.WordCloud(
    background_color='white',
    max_words=200,
    max_font_size=100 ,
    scale=32
)
wc.generate_from_frequencies(freq)
wc.to_file("./wordcloud.png")
plt.imshow(wc)
plt.show()
# This step takes a long time, please wait for a while!
```
